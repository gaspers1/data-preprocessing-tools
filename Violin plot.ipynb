```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

%matplotlib inline
```

```python
# first install library pydataset - pip install pydataset

from pydataset import data
```

```python
dataset = data('PearsonLee')
```

```python
dataset.head()
```

```python
# Set style of seaborn
sns.set(style="whitegrid", palette="pastel", color_codes=True)

# Draw a nested violinplot and split the violins for easier comparison
sv = sns.violinplot(x="par", y="child", hue="chl", data=dataset, inner=None, split=True, palette="Set2", scale="count", linewidth=0.1)

# Show figure
fig = sv.get_figure()
```
